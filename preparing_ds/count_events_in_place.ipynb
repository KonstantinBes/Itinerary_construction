```python
"""
Counts events for every place in KudaGo dataset
Input: two .json lists
with 'dict'`s representing places and events
Outout: new .json list
with 'dict'`s representing places,
for each of place added new field 'events_counted'
"""


import json
from json import JSONDecodeError


location = "msk"    # choose location here (spb, msk)

# In / out directory and files:
direct = "D:/Work/Data_files/working_dir/"
places_file = "places_2_expand_"
events_file = "events_2_expand_"
out_file = "places_events_counted_"

events = []
event_in = {}
places = []

with open(direct + places_file + location + ".json",
           'r', encoding = "utf-8") as pls_f, \
     open(direct + events_file + location + ".json",
           'r', encoding = "utf-8") as evs_f:   
    try:
        events = json.load(evs_f)
        for event in events:
            place = event.get('place')
            if place != None:
                if place['id'] not in event_in:
                    event_in[place['id']] = 1
                else:
                    event_in[place['id']] += 1
        places = json.load(pls_f)
        for place in places:
            if place['id'] not in event_in:
                place['events_count'] = 0
            else:
                events_counted = event_in.pop(place['id'])
                place['events_count'] = events_counted
    except JSONDecodeError:
        print("Input file cannot be read")

if __name__ == __main__:
    # Check the plausability of some output parameters
    max_evs = 0
    max_evs_in = {}
    eventless = 0
    with open(direct + out_file + location + ".json",
              'r', encoding = "utf-8") as inf:
        try:
            places = json.load(inf)
            for place in places:
                if(place['events_count'] > max_evs):
                    max_evs = place['events_count']
                    max_evs_in = place
                if(place['events_count'] == 0):
                    eventless += 1
            print("Max number of events in place: ", max_evs)
            print("Max events in place:\n", max_evs_in)
            print("Places without events: ", eventless)
            if(len(event_in) > 0):
                print("Places not found: ", len(event_in))
                for place in event_in:
                    print(place)
        except JSONDecodeError:
            print("Output file cannot be read")
```
