```python
"""Parser API KudaGo. Full list of places/events in city (spb,msk,...)."""
"""Input: KudaGo.com API url with determined search parameters."""
"""Output: .json file with class 'list' sorted by category[0]:"""
"""[{'id':...,'publication_date':...,'categories':[...]},...]"""


from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup
import json


url_base="https://kudago.com/public-api/v1.4/events/?"
url_fields="&fields=id,publication_date,dates,title,short_title,
!slug,place,description,body_text,location,
!categories,tagline,age_restriction,price,is_free,
!images,favorites_count,comments_count,site_url,tags,participants"
url_expand="&expand=images,place,location,dates,participants"
url_location_time="&location=msk&actual_since=2018-01-01"
url_text_form="&text_format=html"
url_pages="&page_size=20&page="
url_full=url_base+url_fields+url_expand+url_location_time+url_text_form+url_pages
print(url_full+'1')
page=1
# max_pages: spb=528, msk=? found manually for page_size=20


def get_data(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return e
    try:
        data = str(BeautifulSoup(html.read()))
    except AttributeError as e:
        return None
    return data


page_data=''
data=''
with open ('A:\Information\Kostya\Master\Science\Events_msk.json',
           'w', encoding="utf-8") as ouf:
    while (page_data!=None):
        page_data = get_data(url_full+str(page))
        if(type(page_data)!=str):
            break
        if page ==1:
            data=json.loads(page_data[page_data.find('['):-1:1])
            print(type(data[-1]), page)
        else:
            data+=json.loads(page_data[page_data.find('['):-1:1])
            print(type(data[-1]), page)
        page+=1
    category_sorted=sorted(data, key=lambda x: x['categories'][0])
    json.dump(category_sorted, ouf)
    print(page)
with open ('A:\Information\Kostya\Master\Science\Events_msk.json',
           'r', encoding="utf-8") as inf:
    data=json.load(inf)
    print(type(data), data[0], data[-1], len(data))
```
