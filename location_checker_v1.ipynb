```python
"""Place - location check, v_1.0, 25_04_2020"""
"""Checks if places are within the city radius"""
"""Input:"""
"""set of places obtained from API KudaGo.com (parser_2), .json"""
"""Output:"""
"""two .json files: places within the city radius and out of it"""

import re
import json
from json import JSONDecodeError
from geopy import distance


location = "spb"    # Choose city here (spb, msk, ...)
city_coord = {"lat":59.9342802, "lon": 30.3350986}
"""spb: {"lat":59.9342802, "lon": 30.3350986}"""
"""msk: {"lat":-, "lon": -}"""
radius = 50.0    # Border radius from city center, km
"""Recommended: spb: 50.0, msk: 0.0"""


"""Check if object is located within location radius"""
"""type(obj) == 'dict',  coordinates in field: "coords")"""
def obj_within_loc(loc_center,
                   loc_radius, geo_obj):
    geopoint_1 = (loc_center["lat"], loc_center["lon"])
    geopoint_2 = (geo_obj["coords"]["lat"],
                  geo_obj["coords"]["lon"])
    dist_between = distance.distance(geopoint_1,
                                     geopoint_2).km
    if (dist_between) < loc_radius:
        return True
    else:
        return False

"""In / out directorires:"""
in_dir = "D:/Work/Data_files/KudaGo_database/places_2/"
in_file = "places_2_expand_"
out_dir = "D:/Work/Data_files/KudaGo_database/places_3/"
out_file_1 = "places_3_within_"
out_file_2 = "places_3_out_of_"

to_check = []    # all objects (input)
within_loc = []    # ojbects in location (positive)
outof_loc = []    # ojbects out of location (negative)

with open(in_dir + in_file + location + ".json",
           'r', encoding = "utf-8") as inf, \
     open(out_dir + out_file_1 + location + ".json",
           'w', encoding = "utf-8") as ouf_1, \
     open(out_dir + out_file_2 + location + ".json",
           'w', encoding = "utf-8") as ouf_2:
    try:
        to_check = json.load(inf)
        for obj in to_check:
            if obj_within_loc(city_coord, radius, obj):
                within_loc.append(obj)
            else:
                outof_loc.append(obj)
        json.dump(within_loc, ouf_1)
        json.dump(outof_loc, ouf_2)
    except JSONDecodeError:
        print("Input file cannot be read")
        
with open(out_dir + out_file_1 + location + ".json",
          'r', encoding = "utf-8") as inf_1, \
     open(out_dir + out_file_2 + location + ".json",
          'r', encoding = "utf-8") as inf_2:
    try:
        within_loc = json.load(inf_1)
        print("Number of elements in file ", out_file_1 + location)
        print(len(within_loc))
        if(len(within_loc) != 0):
            print("First element:\n", within_loc[0])
            print("Last element:\n", within_loc[-1])
        outof_locat = json.load(inf_2)
        print("Number of elements in file ", out_file_2 + location)
        print(len(outof_loc))
        if(len(outof_loc) != 0):
            print("First element:\n", outof_loc[0])
            print("Last element:\n", outof_loc[-1])
    except JSONDecodeError:
        print("Output file cannot be read")
```
