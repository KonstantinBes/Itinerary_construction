```python
"""Saves in .json and prints list"""
"""of all possible categories in FIFA dataset (ITMO)"""

import json
from json import JSONDecodeError


"""In / out directorires:"""
direct = "D:/Work/Data_files/working_dir/"
in_file_1 = "Saint+Petersburg-finalized"
in_file_2 = "Moscow-finalized"
out_file = "categories_ITMO"


places = []
categories = []

with open(direct + in_file_1 + ".json",
           'r', encoding = "utf-8") as inf_1, \
     open(direct + in_file_2 + ".json",
           'r', encoding = "utf-8") as inf_2, \
     open(direct + out_file + ".json",
           'w', encoding = "utf-8") as ouf:
    try:
        places = json.load(inf_1)
        for place in places:
            if place["categories"] not in categories:
                categories.append(place["categories"])
            if (place["categories"] == ['Nature & Parks', 'Museums & Libraries', 'Sights & Landmarks', 'Others', 'Restaurant']):
                print(place)
        places = json.load(inf_2)
        for place in places:
            if place["categories"] not in categories:
                categories.append(place["categories"])
            if (place["categories"] == ['Nature & Parks', 'Museums & Libraries', 'Sights & Landmarks', 'Others', 'Restaurant']):
                print(place)
    except JSONDecodeError:
        print("Input file cannot be read")
    #json.dump(categories, ouf)

with open(direct + out_file + ".json",
          'r', encoding = "utf-8") as inf:
    try:
        categories = json.load(inf)
        print("Number of elements in file ", out_file)
        print(len(categories))
        if(len(categories) != 0):
            for category in categories:
                print(category)
    except JSONDecodeError:
        print("Output file cannot be read")
```
