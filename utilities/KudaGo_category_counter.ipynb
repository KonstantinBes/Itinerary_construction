```python
"""Prints list of all possible categories in KudaGo dataset"""

import json
from json import JSONDecodeError
from collections import Counter


"""In / out directorires:"""
direct = "D:/Work/Data_files/working_dir/"
in_file_1 = "places_2_expand_msk"
in_file_2 = "places_2_expand_spb"

places = []
categories = []

with open(direct + in_file_1 + ".json",
           'r', encoding = "utf-8") as inf_1, \
     open(direct + in_file_2 + ".json",
           'r', encoding = "utf-8") as inf_2:
    
    try:
        places = json.load(inf_1)
        for place in places:
            if place["categories"] not in categories:
                categories.append(place["categories"])
                
        places = json.load(inf_2)
        for place in places:
            if place["categories"] not in categories:
                categories.append(place["categories"])

    except JSONDecodeError:
        print("Input file cannot be read")

if(len(categories) != 0):
    counter = Counter()
    for category in categories:
        if len(category) > 1:
            for word in category:
                #print([word])
                counter.update([word])
        else:
            #print(category)
            counter.update(category)
    print("Number of categories: ", len(counter))
    for category, number in counter.most_common():
        print(category, number)
```
